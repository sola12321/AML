```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
# read the data
data_noiseless = pd.read_csv("iris.csv").values
data_noisy = {}
j = 1
for i in ["I", "II", "III", "IV", "V"]:
    data_noisy[j] = pd.read_csv("data{}.csv".format(i)).values
    j+=1
```

```python
def PCA(data, num_of_comp, cov_mat, mean):
    """
    PCA using eigenvalues
    data: Input data set, each row represents an observation, each column represents a indiviaul feature.
    num_of_com: number of components left after pca
    cov_mat: convariance matrix used to perform PCA
    mean: mean vector used to perform PCA
    return: data after PCA
    """
    data_centered = data - mean
    eig_values, eig_mat = np.linalg.eig(cov_mat)
    indices = np.argsort(eig_values)[::-1][:num_of_comp]
    data_PCA = np.matmul(data_centered, eig_mat)
    for i in range(len(data_PCA[0])):
        data_PCA[:,i] = data_PCA[:,i] if i in indices else 0
    data_PCA = np.matmul(data_PCA,eig_mat.T) + mean
    return data_PCA
    

def MSE(data, data_PCA):
    """
    Compute the mean squared error between the origin data and the data after PCA
    data: The original data set
    data_PCA: the dataset after PCA 
    """
    return np.mean([np.sum(np.square(data[i]-data_PCA[i])) for i in range(len(data))])
```

```python
res = np.zeros((5, 10))
for data_set_key in range(1, 6):
    for num_of_comp in range(5):
        for j in [0, 5]:
            if j == 0:
                mean = np.mean(data_noiseless, 0)
                data_centered = data_noiseless - mean
                cov_mat = np.cov(data_centered.T)
            else:
                mean = np.mean(data_noisy[data_set_key], 0)
                data_centered = data_noisy[data_set_key] - np.mean(data_noisy[data_set_key], 0)
                cov_mat = np.cov(data_centered.T)
            data_PCA = PCA(data_noisy[data_set_key], num_of_comp, cov_mat, mean)
            res[data_set_key-1][num_of_comp+j] = MSE(data_noiseless, data_PCA)
            if data_set_key == 1 and num_of_comp == 2 and j == 5:
                dataI_PCA_2comps = data_PCA
res           
```

```python
#turn results into pd.dataframe
res_map = {}
i = 0
for key in ["0N","1N","2N","3N","4N","0c","1c","2c","3c","4c"]:
    res_map[key] = res[:,i]
    i += 1
res_df = pd.DataFrame(res_map)
res_df.to_csv("fangwei2-numbers.csv", index=False)

res_map = {}
i = 0
for key in ['"Sepal.Length"','"Sepal.Width"','"Petal.Length"','"Petal.Width"']:
    res_map[key] = dataI_PCA_2comps[:,i]
    i += 1
res_df = pd.DataFrame(res_map)
res_df.to_csv("fangwei2-recon.csv", index=False)
res_df
```
